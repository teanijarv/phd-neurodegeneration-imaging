### partial volume correction PET

```python
amy_prefix = 'fgc'
tau_prefix = 'tgic'
grp_col = 'temporal_meta_tau_asymmetry_group'
covars = ['age', 'gender_baseline_variable']

df = pd.read_csv(os.path.join(df_dir, f'bf2_tau_asymmetry_ad_t_xs_{tau_prefix}_{amy_prefix}.csv'), low_memory=False, index_col=0)

cols = ['sid', 'mid', grp_col] + covars + [col for col in df.columns if col.startswith((amy_prefix, tau_prefix, 'aparc_ct_'))]
df = df[cols].copy()
print(f"N (xs) = {df.shape}")

### plot config

grp_colors = {'LA': '#5785C1FF', 'LA-S': '#ff6961', 'S': '#D3D4D8FF', 'RA-S': '#ff6961', 'RA': '#FBA72AFF'}
statann_kwargs = dict(fontsize=7, color='#494949')
scatter_kwargs = dict(s=15, lw=0.8, alpha=0.8, edgecolor='#494949', color='#9f9f9f')
legend_kwargs = dict(loc='best', title='Tau asymmetry', title_fontsize=7)
strip_kwargs = dict(linewidth=0.8, size=3, alpha=0.8, edgecolor='#494949')
box_kwargs = dict(alpha=0.95, lw=0.8)
pval_kwargs = dict(lw=0.8, fontsize=7, col='#494949', h_coef=0.015, y_coef=0.1)

label_dict = {f'{tau_prefix}_temporal_meta_LI': 'Temporal tau laterality', f'{tau_prefix}_cho_com_I_II_LI': 'Braak I-II tau laterality', 
              f'{tau_prefix}_cho_com_III_IV_LI': 'Braak III-IV tau laterality', f'{tau_prefix}_cho_com_V_VI_LI': 'Braak V-VI tau laterality', 
              f'{amy_prefix}_temporal_meta_LI': 'Temporal Aβ laterality', f'{amy_prefix}_cho_com_I_II_LI': 'Braak I-II Aβ laterality', 
              f'{amy_prefix}_cho_com_III_IV_LI': 'Braak III-IV Aβ laterality', f'{amy_prefix}_cho_com_V_VI_LI': 'Braak V-VI Aβ laterality',
              f'{tau_prefix}_global_LI': 'Global tau laterality', f'{amy_prefix}_global_LI': 'Global Aβ laterality'}

fig, axs = plt.subplots(1, 4, figsize=(180/25.4, 50/25.4))

# subplot for regression between Ab vs tau lateralities

for i, roi in enumerate(['global_LI', 'cho_com_I_II_LI', 'cho_com_III_IV_LI', 'cho_com_V_VI_LI']):
    y = f'{tau_prefix}_{roi}'
    x = f'{amy_prefix}_{roi}'

    df_ = df.dropna(subset=[x, y])
    

    df_ = df_.loc[abs(df_[x]) <= 50] # one outlier removed with Ab laterality over 50 at Braak I-II
    _, axs[i], model = plot_regression(df=df_, x=x, y=y, covars=['age', 'gender_baseline_variable'], standardise=True,
                                ax=axs[i], xlabel=label_dict[x], ylabel=label_dict[y], 
                                text_kwargs=statann_kwargs, scatter_kwargs=scatter_kwargs)
    # print(model.summary())


plt.suptitle("Sensitivity analysis: partial volume corrected PET SUVR", fontsize=7)
plt.tight_layout()
plt.savefig(os.path.join(plot_dir, f'suppl_sens_partialvolcor_xs_laterality.pdf'), format='pdf', dpi=300)
plt.show()
```

### Schaefer parcellation

```python
set_plot_style(dpi=150)
statann_kwargs = dict(fontsize=7, color='#494949')
scatter_kwargs = dict(s=15, lw=0.8, alpha=0.8, edgecolor='#494949', color='#9f9f9f')

warnings.simplefilter(action='ignore')

def load_mat_data_to_dataframe(base_dir, df, id_col, var='data_var'):
    row_data = []
    for i, row in df.iterrows():
        if pd.notna(row[id_col]):
            data_dir = os.path.join(base_dir, row[id_col], 'HB')
            if os.path.exists(data_dir):
                mat_files = [f for f in os.listdir(data_dir) if f.endswith('.mat')]
                for mat_file in mat_files:
                    file_path = os.path.join(data_dir, mat_file)
                    mat_data = scipy.io.loadmat(file_path)
                    if var in mat_data:
                        data = mat_data[var][:, 0]
                        row_data.append({id_col: row[id_col], var: data})
    df_merged = pd.merge(df, pd.DataFrame(row_data), on=id_col, how='left')
    return df_merged

def calculate_hemisphere_averages(data):
    if isinstance(data, (list, np.ndarray)) and len(data) == 400:
        left_avg = np.mean(data[:200])
        right_avg = np.mean(data[200:])
        return left_avg, right_avg
    else:
        return np.nan, np.nan
    
amy_prefix = 'fnc'
tau_prefix = 'tnic'

df = pd.read_csv(os.path.join(df_dir, f'bf2_tau_asymmetry_ad_t_xs_{tau_prefix}_{amy_prefix}.csv'), low_memory=False, index_col=0)

apet_dir = os.path.join(data_dir, 'schaefer', 'amypet')
tpet_dir = os.path.join(data_dir, 'schaefer', 'taupet')

_, sch_labels = fetch_schaefer_atlas(400)
sch_labels = list(sch_labels.keys())

ROIs = get_sch_rois()

def convert_array(x):
    if isinstance(x, np.ndarray):
        return x.tolist()
    return [np.nan] * len(sch_labels)

df_bf2_ap = load_mat_data_to_dataframe(apet_dir, df, id_col='flutemetamol_pet_date__index', var='AP')
df_ap_data = pd.DataFrame([convert_array(x) for x in df_bf2_ap['AP']], columns=sch_labels)
df_ap_data.columns = df_ap_data.columns.str.replace('7Networks', 'fnc')
df_ap_data = pd.concat([df_bf2_ap[['mid', 'proc_flutemetamol_pet_date']], df_ap_data], axis=1).dropna().reset_index(drop=True)

df_bf2_tp = load_mat_data_to_dataframe(tpet_dir, df, id_col='tau_pet_date__index', var='TP')
df_tp_data = pd.DataFrame([convert_array(x) for x in df_bf2_tp['TP']], columns=sch_labels)
df_tp_data.columns = df_tp_data.columns.str.replace('7Networks', 'tnic')
df_tp_data = pd.concat([df_bf2_tp[['mid', 'proc_tau_pet_date']], df_tp_data], axis=1).dropna().reset_index(drop=True)

df_sch_data = pd.merge(df_ap_data, df_tp_data, on='mid', how='inner')

def assign_value_to_regions(selected_regions, value, sch_labels):
    # Initialize array with NaN values
    result = np.full(len(sch_labels), np.nan)
    
    # Create a mapping of regions to their indices in sch_labels
    for region in selected_regions:
        # Find index of the region in sch_labels and assign the value
        if region in sch_labels:
            idx = sch_labels.index(region)
            result[idx] = value
            
    # Convert to pandas Series with sch_labels as index
    return pd.Series(data=result, index=sch_labels)

sch_labels_ = [col.split('_', 1)[-1] for col in sch_labels]

label_dict = {f'{tau_prefix}_temporal_meta_LI': 'Temporal tau laterality', f'{amy_prefix}_temporal_meta_LI': 'Temporal Aβ laterality',
              f'{tau_prefix}_global_LI': 'Global tau laterality', f'{amy_prefix}_global_LI': 'Global Aβ laterality'}

fig, axs = plt.subplots(1, 2, figsize=(90/25.4, 50/25.4))

for i, (ROI_name, ROI_regions) in enumerate(ROIs.items()):
    if ROI_name == 'global':
        ROI_regions = [reg.split('_', 1)[-1] for reg in ROI_regions]
    ROI_regions_ = [reg.split('_', 1)[-1] for reg in ROI_regions]

    # amyloid-PET
    df_sch_data[f'fnc_{ROI_name}_left'] = compute_roi_amyloid(df_sch_data, ROI_regions_, ['_LH_'], pet_prefix='fnc_')
    df_sch_data[f'fnc_{ROI_name}_right'] = compute_roi_amyloid(df_sch_data, ROI_regions_, ['_RH_'], pet_prefix='fnc_')
    df_sch_data[f'fnc_{ROI_name}'] = compute_roi_amyloid(df_sch_data, ROI_regions_, ['_LH_', '_RH_'], pet_prefix='fnc_')
    # amyloid laterality index
    df_sch_data[f'fnc_{ROI_name}_LI'] = df_sch_data.apply(lambda row: laterality_index(
        row[f'fnc_{ROI_name}_left'], row[f'fnc_{ROI_name}_right']), axis=1)
    
    # tau-PET
    df_sch_data[f'tnic_{ROI_name}_left'] = compute_roi_tau(df_sch_data, ROI_regions_, ['_LH_'], pet_prefix='tnic_')
    df_sch_data[f'tnic_{ROI_name}_right'] = compute_roi_tau(df_sch_data, ROI_regions_, ['_RH_'], pet_prefix='tnic_')
    df_sch_data[f'tnic_{ROI_name}'] = compute_roi_tau(df_sch_data, ROI_regions_, ['_LH_', '_RH_'], pet_prefix='tnic_')
    # tau laterality index
    df_sch_data[f'tnic_{ROI_name}_LI'] = df_sch_data.apply(lambda row: laterality_index(
        row[f'tnic_{ROI_name}_left'], row[f'tnic_{ROI_name}_right']), axis=1)
    
    x = f'fnc_{ROI_name}_LI'
    y = f'tnic_{ROI_name}_LI'

    _,axs[i],mdl = plot_regression(df=df_sch_data, x=x, y=y, xlabel=label_dict[x], ylabel=label_dict[y],
                        ax=axs[i], standardise=True,
                        text_kwargs=statann_kwargs, scatter_kwargs=scatter_kwargs)
    axs[i].set_ylim([-42, 42])
    # axs[i].set_xlim([-10, 10])

    surf_args = dict(data=assign_value_to_regions(ROI_regions, mdl.params[-1], sch_labels_), atlas='schaefer400', cmap='Reds', cmap_range=(0.3, 0.7))

    fig_surf = plot_surface(**surf_args, plot_kwargs=dict(nan_color=(0.8, 0.8, 0.8, 1), zoom=1.35, transparent_bg=False, scale=1))
    _ = plot_surface(**surf_args, output=os.path.join(surfplot_dir, f'surf_sch_{x}_with_{y}_beta.png'))
    display(fig_surf)
    
plt.suptitle("Sensitivity analysis: Schaefer-400 atlas parcellation", fontsize=7)
plt.tight_layout()
plt.savefig(os.path.join(plot_dir, f'suppl_sens_schaefer_xs_laterality.pdf'), format='pdf', dpi=300)
plt.show()
```
